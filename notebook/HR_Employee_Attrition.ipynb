```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("HREmployeeAttrition.csv")
```

```python
# Show Header of the data
df.head()
```

```python
# EDA and check nulls
df.info()
df.isnull().sum()
```

```python
df.describe()
```

```python
# Remove unneeded columns
used_columns = ['DistanceFromHome','Education','EnvironmentSatisfaction','JobRole','MonthlyIncome','YearsAtCompany','Attrition']
df = df[used_columns]
df
```

```python
# Question 1: 1-How is the distance from home by job role effects on the attrition of the employees?
# Analyis
df.groupby(['JobRole','Attrition'])['DistanceFromHome'].mean().unstack()
```

The data reveals that the average distance from home varies by job role and attrition status. Notably, employees in roles such as Healthcare Representative and Human Resources who left the company tended to live significantly farther from work (17.7 km and 13.4 km respectively) than those who stayed (9.2 km and 6.6 km). This suggests that commuting distance may be a contributing factor to attrition in these roles.

Conversely, for some job roles such as Manufacturing Director and Research Director, those who stayed actually had slightly longer average commutes than those who left, indicating that distance may not be a strong attrition factor for every position.

This highlights the importance of considering job-specific patterns when analyzing workforce retention. In roles where employees are more sensitive to commuting time, offering flexible work arrangements or relocation support may improve retention.

```python
# Question 2: 2-How the average monthly income by education effects on attrition of the employees?
# Analyis
df.groupby(['Education','Attrition'])['MonthlyIncome'].mean().unstack()
```

The data confirms a strong correlation between education level and income ....
higher education leads to higher pay across both retained and exited employees, Expected.

However, those who left consistently earned less than those who stayed at every education level. This may indicate that salary dissatisfaction, rather than education itself, contributes to higher attrition risk.

It is recommended that the company periodically monitors and benchmarks the average income of current employees. If it is found that certain groups like by education level are earning at levels similar to those who previously left the company, proactive adjustments may help improve retention and reduce future attrition risk

```python
# Question 3: 3-How the average monthly income by education effects on attrition of the employees?
# Analyis
colors = {'Yes': 'red', 'No': 'green'}

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='YearsAtCompany', y='MonthlyIncome', hue='Attrition', palette={'Yes': 'red', 'No': 'green'}, alpha=0.6)

plt.title('Income vs. Years at Company by Attrition')
plt.show()
```

The data reflects a strong concentration of employee exits within the first few years at the company, especially among lower-income brackets. Additionally, some employees with long tenure remain at similar income levels, which might pose a risk of dissatisfaction. This indicates a potential need for better salary progression and engagement strategies to retain experienced talent.

We recommend monitoring the average income progression across tenure brackets. Early-stage employees with low income are at a higher attrition risk. Addressing compensation stagnation for mid- to long-term employees could also enhance retention.
